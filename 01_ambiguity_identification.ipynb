```python
import pandas as pd

ambiguity_taxonomy = pd.read_csv('ambiguity_taxonomy.csv')
ambiguity_examples = pd.read_csv('ambiguity_examples.csv')
scenarios = pd.read_csv('all_cases_tasks.csv')

# scenario and task extraction example
scenario = scenarios['Scenario'][0] #first scenario
task = scenarios['Task with 3 ambiguities'][0] # high ambiguity task of first scenario
```
